## Q 5. Find the net revenue generated by the E-Commerce company over all orders placed. (6 marks)

### Ans 5

```python
#creating a user defined function(UDF) that calculatesthe revenue generated by Ecom per transaction and returns sum of the items
def rev_Ecom(disc_price):
    Ecom_rev=0
    for i in  disc_price:
            if i>600:
                Ecom_rev=Ecom_rev+i*0.25
            elif (i >350) and (i<=600):
                Ecom_rev=Ecom_rev+i*0.15
            elif (i >100) and (i<=350):
                Ecom_rev=Ecom_rev+i*0.10
            else :
                Ecom_rev=Ecom_rev+i*0.05
    return(print("Total revenue generated by the E-commerce company of all order placed is ₹{}".format(round(Ecom_rev,2))))
#Converting the column -discounted_price to list as the user defined function takes list as input
disc_price_list=ECom_Data['discounted_price'].tolist()
#Executing the UDF
rev_Ecom(disc_price_list)
```

## Q6.Calculate the Net BrandRevenue for each Brand and list the top 14 Brand having maximum revenue in descending order (6 marks)
#### Hint: Net BrandRevenue is BrandRevenue which is generated after all the deductions. Also, BrandRevenue is different from the E-Commerce company's net revenue.

### Ans 6

```python
    #converting discounted price column into list
    disc_price_list=ECom_Data['discounted_price'].tolist()
    #Starting with a blank list
    Ecom_revenue=[]
    Net_Brand_revenue=[]
    #creating the two calculated fields
    for i in  disc_price_list:
            if i>600:
                Ecom_rev=i*0.25
                Ecom_revenue.append(Ecom_rev)
                Net_Brand_rev=(i-Ecom_rev)
                Net_Brand_revenue.append(Net_Brand_rev)
            elif (i >350) and (i<=600):
                Ecom_rev=i*0.15
                Ecom_revenue.append(Ecom_rev)
                Net_Brand_rev=(i-Ecom_rev)
                Net_Brand_revenue.append(Net_Brand_rev)
            elif (i >100) and (i<=350):
                Ecom_rev=i*0.10
                Ecom_revenue.append(Ecom_rev)
                Net_Brand_rev=(i-Ecom_rev)
                Net_Brand_revenue.append(Net_Brand_rev)
            else :
                Ecom_rev=i*0.05
                Ecom_revenue.append(Ecom_rev)
                Net_Brand_rev=(i-Ecom_rev)
                Net_Brand_revenue.append(Net_Brand_rev)
    
ECom_Data_new=ECom_Data.copy()
#Adding the calculated fields to the datasets and creating a new dataset so as to keep the original unchanged
ECom_Data_new['Ecom_Margin']=Ecom_revenue
ECom_Data_new['Net_Brand_Revenue']=Net_Brand_revenue

#Display the top 14
ECom_Data_new.groupby('Brand')['Net_Brand_Revenue'].sum().sort_values(ascending=False).head(14)
    
```
